## KOSIS 차종별 시군구별 자동차주행거리

**통계산정기준은 기준년도 실제 주행거리가 아닌 기준년도 자동차검사를 받은 차량의 이전검사로부터 최근검사까지의 평균 주행거리임.**

1일 평균주행거리

1) 1일 평균주행거리
    주행거리를 운행일수로 나눈 값임
2) 주행거리
    최근 검사 주행거리에서 이전 검사 주행거리를 뺀 거리
3) 운행일수
    최근 검사일에서 이전 검사일 사이 일수

연간 주행거리

1) 1일 평균주행거리×365일×자동차등록대수(모집단)
2) 자동차등록대수는 6월기준

4) 행정구역별 > 미추홀구

* '인천광역시 남구 명칭변경에 관한 법률'에 따라 남구의 이름이 미추홀구로 개칭(2018.7.1. 시행)

5) 행정구역별 > 여주시

여주군 → 여주시 (2013.9월)

```python
import pandas as pd
import koreanize_matplotlib
import plotly.express as px
from glob import glob
import matplotlib.pyplot as plt
```

```python
file_path = glob("project/*자동차*")
file_path
```

```python
df.columns = ["시도명", "시군구", "연도", "주행거리항목", "자동차용도별", "차종별", "데이터"]
df.columns
```

```python
df["주행거리항목"].unique()
```

```python
df_year = df[(df["시군구"]=="소계") & (df["주행거리항목"]=="연간 주행거리 (천㎞)")]
df_day = df[(df["시군구"]=="소계") & (df["주행거리항목"]=="1일 평균주행거리 (㎞/대)")]
```

```python
df_year
```

```python
# 시도별 차종별 일평균 주행거리
px.histogram(df_day, x="시도명", y="데이터", color="차종별")
```

```python
# 시도별 차종별 연간 주행거리
px.histogram(df_year, x="시도명", y="데이터", color="차종별")
```

```python
px.histogram(df_year, x="연도", y="데이터", color="차종별")
```

시도별 주행거리
* 연간 주행거리 : 1일 평균주행거리×365일×자동차등록대수(모집단) * 자동차등록대수는 6월기준
* 일평균 주행거리에서 다른 시도보다 경기도가 높다.
* 연간 주행거리에서 최고로 나타나는 것으로 보아 자동차등록대수가 다른 시도보다 많음을 알 수 있다.

업종을 가져오자

```python
file_path = glob("project/*업종*")[1]
file_path
```

```python
df_cfac = pd.read_csv(file_path, encoding="cp949")
```

```python
df_cfac["연도"] = df_cfac["시점"].str.split(".", expand=True)[0]
df_cfac["연도"] = df_cfac["연도"].astype("int32")
```

```python
df_cfac["데이터"] = df_cfac["데이터"].str.replace("-", "0")
df_cfac["데이터"] = df_cfac["데이터"].astype("int32")
```

```python
kb_top5 = df_cfac[df_cfac["시도별"]=="경북"].groupby(["업종별"])["데이터"].sum().sort_values(ascending=False).head()
kb_top5
```

```python
cn_top5 = df_cfac[df_cfac["시도별"]=="충남"].groupby(["업종별"])["데이터"].sum().sort_values(ascending=False).head()
cn_top5
```

```python
jn_top5 = df_cfac[df_cfac["시도별"]=="전남"].groupby(["업종별"])["데이터"].sum().sort_values(ascending=False).head()
jn_top5
```

```python
kk_top5 = df_cfac[df_cfac["시도별"]=="경기"].groupby(["업종별"])["데이터"].sum().sort_values(ascending=False).head()
kk_top5
```

```python
# 경기와 비교 했을 때 경북에만 있는 업종
list(set(kb_top5.index)-set(kk_top5.index))
```

```python
# 경기와 비교 했을 때 충남에만 있는 업종
list(set(cn_top5.index)-set(kk_top5.index))
```

```python
# 경기와 비교 했을 때 전남에만 있는 업종
list(set(jn_top5.index)-set(kk_top5.index))
```

'자동차 및 트레일러 제조업', '식료품 제조업', '섬유제품 제조업;의복제외', '비금속 광물 제품 제조업', <br>

# 정리


* 비산먼지<br>
비산먼지를 가장 많이 배출하는 경기도가 연간 주행거리가 가장 많은 것을 볼 수 있다.<br>
일평균 주행거리에 비해 타 시도 보다 높은 것을 봐서 자동차 등록 대수가 많은 것을 알 수 있다.<br>
<br>
화물차의 주행거리도 높은 것을 봐서 경기도의 비산먼지 배출량이 높은 이유를 짐작할 수 있다.<br>
연간 주행거리는 계속 늘어나는 중이다.<br>
<br>
* 제조업 연소<br>
<br>
공장수/업종별 소유형태로 지역별 차이를 파악하려 했으나, 제조업 연소 배출을 높이는 이유를 파악하기 힘들었다.<br>
하지만, 제조업부문 업종별 에너지소비량 데이터를 통해 추론해 볼 수 있었다.<br>
<br>
석탄 : 1차 철강제조업, 시멘트제조업<br>
석유 : 석유정제품제조업, 비료농약및살균살충제제조업 에서 많이 소비한다.<br>
<br>
국가미세먼지정보센터는 제조업 연소 배출원별 연료별 배출계수를 석유보다 석탄에 더 크게 부과하고 있다.<br>
따라서 석탄에 더 중점을 두고, 경북, 충남, 전남에 1차 철강 제조업이 더 많은건 아닐까? 추측해 볼 수 있었고,<br>
실제 경상북도, 전라남도, 충청남도 순으로 철강 산업의 출하액이 높았다.
